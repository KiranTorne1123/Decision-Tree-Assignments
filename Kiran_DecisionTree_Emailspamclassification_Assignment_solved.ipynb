```python
#drop the Email.No column as it serves no purpose in this dataset
# we remove Email no column from the dataset so we give df.columns[1:] and we create 2 variables 
# for x we take all rows and all columns except the last one, so [:-1]
# for y we take only the last column so [-1]
# create input data as x input
x = df[df.columns[1:-1]]
# create input data as y input
y = df[df.columns[-1]]
```

```python
# we have our x and y variable ready.
# now we import new libraries to encode the Email column
from sklearn import preprocessing
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
# now we create training and testing data for our model
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=69)
```

```python
# now we check our testing and training variables
x_train.head(5)
```

```python
#now we import decision tree classfier
from sklearn.tree import DecisionTreeClassifier
```

```python
# Creating new model
model_kiran = DecisionTreeClassifier()
```

```python
model_kiran
```

```python
# Our training and testing x, y variables are ready.
# now we fit them in a model
model_kiran.fit(x_train,y_train)
```

```python
# now the training data is fed to our model
# Now testing new Model prediction by providing Test Data set
y_kiran=model_kiran.predict(x_test)
```

```python
# this is our variable which has learned the test data
y_kiran
```

```python
# to check accuracy we need to import another library
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Compare predictions with testing data given
# this will only give accuracy
accuracy = accuracy_score(y_test, y_kiran)
print(accuracy)
```

```python
# Generate a detailed classification report
report = classification_report(y_test, y_kiran)
print(report)
```

```python
#  so finally we get an accuracy of 92% when we keep test_size as 80/20
# if we change this to 70/30 then
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=69)
model_kiran.fit(x_train,y_train)
y_kiran=model_kiran.predict(x_test)
accuracy = accuracy_score(y_test, y_kiran)
print(accuracy)
report = classification_report(y_test, y_kiran)
print(report)
```

```python
# here we see that 70/30 splitting gives more better accuracy of 93%.
# thus we conclude that 70/30 split gives better accuracy for this dataset.
```
